```python
import re
import nltk
import string
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from unidecode import unidecode

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Inicializações
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('portuguese'))

def preprocess_text(text):
    def substituir_dominios(texto):
        def extrair_dominio(url):
            dominio = re.sub(r'^https?://(?:www\.)?|www\.', '', url)
            dominio = re.split(r'[/?#]', dominio)[0]
            return dominio.split('.')[0]
        return re.sub(r'https?://(?:www\.)?\S+|www\.\S+', lambda m: extrair_dominio(m.group(0)), texto)

    # Substituir domínios
    text = substituir_dominios(text)

    # Minúsculas e remoção de acentos
    text = unidecode(text.lower())

    # Remover pontuações
    text = text.translate(str.maketrans('', '', string.punctuation))

    # Substituir emojis e risadas repetidas
    text = re.sub(r'([\U00010000-\U0010FFFF])\1+', r'\1', text)
    text = re.sub(r'(ha){2,}', 'haha', text, flags=re.IGNORECASE)
    text = re.sub(r'(ka){2,}', 'kaka', text, flags=re.IGNORECASE)
    text = re.sub(r'k{2,}', 'kk', text, flags=re.IGNORECASE)

    # Reduzir espaços
    text = re.sub(r'\s+', ' ', text).strip()

    # Remover stopwords
    palavras_filtradas = [word for word in text.split() if word not in stop_words]

    # Lematização
    palavras_filtradas = [lemmatizer.lemmatize(word) for word in palavras_filtradas]

    return ' '.join(palavras_filtradas)
```

Preparando os textos com BoW e extraindo o vocabulário, para depois aplicar LDA ou outro modelo de tópicos

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import numpy as np

vectorizer = CountVectorizer(max_features=1000) # Criando a matriz BoW
bow_matrix = vectorizer.fit_transform(texts) # Aplicando a vetorização
vocab = vectorizer.get_feature_names_out() # Obtendo o vocabulário
```

###**1. LDA** (Latent Dirichlet Allocation)

Modelo estatístico que descobre automaticamente os **tópicos principais** de um conjunto de textos, assumindo que **cada documento trata de vários assuntos ao mesmo tempo**.

```python
lda = LatentDirichletAllocation(n_components=10, random_state=42)
lda.fit(bow_matrix)

def print_top_words(model, feature_names, n_top_words=10):
    for topic_idx, topic in enumerate(model.components_):
        print(f"Tópico #{topic_idx}:")
        print(" ".join([feature_names[i] for i in topic.argsort()[:-n_top_words - 1:-1]]))

print("🔷 Tópicos - LDA:")
print_top_words(lda, vocab)
```

```python
import numpy as np
import pandas as pd

texts = df_textos['preprocessed'].astype(str).tolist()
topic_distributions = lda.transform(bow_matrix)

df_lda_resultados = pd.DataFrame({
    "Texto": texts,
    "Tópico_Principal": np.argmax(topic_distributions, axis=1),
    "Probabilidade": np.max(topic_distributions, axis=1).round(3),
})

print("📄 Classificação dos documentos por LDA:")
df_lda_resultados.head(10)
```

```python
print("\n📊 Quantidade de documentos atribuídos a cada tópico (LDA):")
print(df_lda_resultados["Tópico_Principal"].value_counts().sort_index())
```

### **2. GSDMM** (Gibbs Sampling Dirichlet Multinomial Mixture)

**Algoritmo de clusterização** projetado para textos **curtos**, que agrupa documentos semelhantes e assume que **cada texto pertence a um único tópico principal**.

```python
from scipy.special import logsumexp

class GSDMM:
    def __init__(self, n_topics=10, alpha=0.1, beta=0.1, n_iter=30):
        # Inicializa os parâmetros do modelo
        self.n_topics = n_topics
        self.alpha = alpha
        self.beta = beta
        self.n_iter = n_iter
        self.eps = 1e-12 # Para evitar log(0) e divisões por zero

    def fit(self, bow_matrix, vocab):
        # Converte para array denso se estiver em formato esparso
        if hasattr(bow_matrix, "toarray"):
            bow_matrix = bow_matrix.toarray()

        n_docs, n_words = bow_matrix.shape

        # Inicializa contadores de palavras por tópico e documentos por tópico
        self.word_topic_counts = np.zeros((n_words, self.n_topics)) + self.beta
        self.topic_counts = np.zeros(self.n_topics) + self.alpha

        # Atribui tópicos iniciais aleatórios a cada documento
        self.doc_topics = np.random.randint(0, self.n_topics, n_docs)

        # Amostragem de Gibbs para redistribuir os documentos entre os tópicos
        for _ in range(self.n_iter):
            for i in range(n_docs):
                doc = bow_matrix[i]
                current_topic = self.doc_topics[i]

                # Remove o documento atual do tópico antigo
                self.word_topic_counts[:, current_topic] -= doc
                self.topic_counts[current_topic] -= 1

                # Garante que contagens não fiquem negativas
                np.maximum(self.word_topic_counts, self.eps, out=self.word_topic_counts)
                np.maximum(self.topic_counts, self.eps, out=self.topic_counts)

                # Calcula a distribuição de probabilidade sobre os tópicos
                log_probs = np.log(self.topic_counts + self.eps)
                log_probs += np.sum(np.log(self.word_topic_counts.T + self.eps) * doc, axis=1)

                # Normaliza e converte de volta ao espaço de probabilidade
                log_probs -= logsumexp(log_probs)
                probs = np.exp(log_probs)

                # Amostra um novo tópico com base nas probabilidades
                try:
                    new_topic = np.random.choice(self.n_topics, p=probs)
                except:
                    new_topic = np.argmax(probs)

                # Atualiza os contadores com o novo tópico atribuído
                self.word_topic_counts[:, new_topic] += doc
                self.topic_counts[new_topic] += 1
                self.doc_topics[i] = new_topic

        self.vocab = vocab # Armazena o vocabulário para uso posterior

    def get_topics(self, top_n=10):
        # Retorna as top_n palavras mais frequentes por tópico
        topics = {}
        for k in range(self.n_topics):
            top_words_idx = np.argsort(-self.word_topic_counts[:, k])[:top_n]
            topics[k] = [self.vocab[i] for i in top_words_idx]
        return topics

# Instancia e treina o modelo
gsdmm = GSDMM(n_topics=5)
gsdmm.fit(bow_matrix, vocab)

# Exibe os tópicos extraídos
print("🔷 Tópicos - GSDMM:")
for topico, palavras in gsdmm.get_topics().items():
    print(f"Tópico {topico}: {', '.join(palavras)}")
```

```python
df_gsdmm_resultados = pd.DataFrame({
    "Texto": texts,
    "Tópico_Associado": gsdmm.doc_topics
})

print("📄 Classificação dos documentos por GSDMM:")
df_gsdmm_resultados.head(10)
```

```python
# Mostra a contagem de documentos por tópico
print("\n📊 Quantidade de documentos atribuídos a cada tópico (GSDMM):")
print(df_gsdmm_resultados["Tópico_Associado"].value_counts().sort_index())
```

###**3. PTM** (usando LDA como aproximação)

PTM não é um modelo específico, mas uma categoria de modelos que incluem o LDA, usados para descobrir temas latentes em coleções de documentos por meio de estatísticas e probabilidades.

Para representar o modelo PTM (Probabilistic Topic Modeling), foi utilizado o algoritmo LDA com diferentes hiperparâmetros, considerando que o LDA é um dos principais representantes dessa classe de modelos. Essa abordagem visa explorar como a variação nos parâmetros afeta a formação dos tópicos, mantendo a coerência com a proposta de avaliação experimental entre métodos probabilísticos e não probabilísticos.

```python
ptm = LatentDirichletAllocation(n_components=10, learning_method='online', max_iter=20, random_state=123)
ptm.fit(bow_matrix)
print("🔷 Tópicos - PTM:")
print_top_words(ptm, vocab)
```

```python
# Classificação dos documentos com o modelo PTM
topic_distributions_ptm = ptm.transform(bow_matrix)

df_ptm_resultados = pd.DataFrame({
    "Texto": texts,
    "Tópico_Principal": np.argmax(topic_distributions_ptm, axis=1),
    "Probabilidade": np.max(topic_distributions_ptm, axis=1).round(3),
})

print("📄 Classificação dos documentos por PTM:")
df_ptm_resultados.head(10)
```

```python
# Mostrar a contagem de documentos por tópico
print("\n📊 Quantidade de documentos atribuídos a cada tópico (PTM):")
print(df_ptm_resultados["Tópico_Principal"].value_counts().sort_index())
```

### **4. BERTopic** (Bidirectional Encoder Representations Topic)

BERTopic é uma técnica de modelagem de tópicos que usa **representações semânticas** de palavras baseadas em BERT (ou outros modelos transformers) para **detectar tópicos com mais precisão**, especialmente em textos curtos ou informais.

```python
from bertopic import BERTopic
from sentence_transformers import SentenceTransformer

# Carregar modelo de embeddings
embedding_model = SentenceTransformer("paraphrase-multilingual-MiniLM-L12-v2")

# Criar e treinar BERTopic
bertopic_model = BERTopic(
    embedding_model=embedding_model,
    language="portuguese",
    top_n_words=10,
    min_topic_size=15
)

topics, probs = bertopic_model.fit_transform(texts[:5000])
```

```python
# Mostrar os tópicos
print("🔷 Tópicos - BERTopic:")
bertopic_model.get_topic_info().head(10)
```

```python
# 2️⃣ Classificação dos documentos
df_bertopic_resultados = pd.DataFrame({
    "Texto": texts[:5000],
    "Tópico_Principal": topics,
    "Probabilidade": probs
})

print("\n📄 Classificação dos documentos por BERTopic:")
df_bertopic_resultados.head(10)
```

```python
# 3️⃣ Quantidade de documentos por tópico
print("\n📊 Quantidade de documentos atribuídos aos 20 tópicos mais frequentes (BERTopic):")
print(df_bertopic_resultados["Tópico_Principal"].value_counts().head(20))
```
